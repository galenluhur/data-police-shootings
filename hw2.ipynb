```python
import string

punctuation_to_keep = ','

def remove_punctuation_except_commas(text):
    if pd.notna(text):
        return ''.join(char if char.isalnum() or char in punctuation_to_keep else ' ' for char in text)
    return text

df2['Author'] = df2['Author'].apply(remove_punctuation_except_commas)

df2['Publisher'] = df2['Publisher'].apply(remove_punctuation_except_commas)

df2
# I used chatGPT for help. thank you chatGPT
```

```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
pima = pd.read_csv("diabetes.csv", header=None, names=col_names)
pima = pima.iloc[1:]

pima.head()
```

```python
#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Assuming clf is your DecisionTreeClassifier instance

# Plot the decision tree
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=feature_cols, class_names=['0', '1'], rounded=True)
plt.savefig('diabetes.png')
plt.show()

#I used chat to change it to only use tree and matplotlib
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# i used chat
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Assuming X_train and X_test are your training and testing feature matrices
# Assuming y_train and y_test are your training and testing labels

# Create a DecisionTreeClassifier with pruning parameters
clf_pruned = DecisionTreeClassifier(max_depth=3, min_samples_split=5, min_samples_leaf=2, random_state=42)
X_train_pruned = X_train.drop('glucose', axis=1)
X_test_pruned = X_test.drop('glucose', axis=1)
clf_pruned.fit(X_train_pruned, y_train)

# Plot the pruned decision tree
plt.figure(figsize=(12, 8))
tree.plot_tree(clf_pruned, filled=True, feature_names=X_train_pruned.columns, class_names=['0', '1'], rounded=True)
plt.savefig('diabetes_pruned.png')
plt.show()

# i used chat
```

```python
from sklearn.metrics import accuracy_score

# Assuming clf is your original decision tree classifier
y_pred_original = clf.predict(X_test)
accuracy_original = accuracy_score(y_test, y_pred_original)

# Assuming clf_pruned is your sugar-free decision tree classifier
y_pred_pruned = clf_pruned.predict(X_test_pruned)
accuracy_pruned = accuracy_score(y_test, y_pred_pruned)

print("Accuracy with original tree (including glucose):", accuracy_original)
print("Accuracy with sugar-free decision tree:", accuracy_pruned)

# i used chat
```
