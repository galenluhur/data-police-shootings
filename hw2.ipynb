```python
import pandas as pd 
df = pd.read_csv("flickr.csv")
df.head(30) #2
```

```python
selected = ['Identifier', 'Publisher', 'Author', 'Title', 'Flickr URL']
df2 = df[selected]
df2 #3
```

```python
df2.columns = df2.columns.str.replace(' ', '_')
df2 #4
```

```python
import string

punctuation_to_keep = ','

def remove_punctuation_except_commas(text):
    if pd.notna(text):
        return ''.join(char if char.isalnum() or char in punctuation_to_keep else ' ' for char in text)
    return text

df2['Author'] = df2['Author'].apply(remove_punctuation_except_commas)

df2['Publisher'] = df2['Publisher'].apply(remove_punctuation_except_commas)

df2
# I used chatGPT for help. thank you chatGPT
```

```python
df2['Flickr_URL'] = df2['Flickr_URL'].str.replace('http://', 'https://')
df2 #6
```

```python
df2['Author'].fillna('unknown', inplace=True)
df2['Publisher'].fillna('unknown', inplace=True)
df2 #7
```
